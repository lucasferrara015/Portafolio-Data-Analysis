```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Configuración de gráficos
sns.set_theme(style="whitegrid")

# Cargar el archivo con su nombre verbatim
filename = "Encuesta campo laboral en la comunicación digital - Respuestas de formulario 1.csv"
df = pd.read_csv(filename)

# Renombrar columnas con nombres cortos y limpios para el código
df.columns = [
    'timestamp', 'orientacion', 'experiencia', 'rol', 'tipo_organizacion',
    'habilidades_importantes', 'comodidad_digital', 'desarrollo_habilidades',
    'p7_abierta', 'aspecto_satisfactorio', 'p9_desafios'
]

# Limpieza de espacios en blanco en cadenas de texto
for col in ['experiencia', 'rol', 'tipo_organizacion']:
    df[col] = df[col].str.strip()

print(f"Datos cargados. Total de encuestados: {len(df)}")
```

```python
# Filtramos las respuestas vacías o respuestas nulas (como guiones "-")
respuestas_p7 = df['p7_abierta'].dropna()
respuestas_p7 = respuestas_p7[respuestas_p7.str.len() > 2]

print(f"Total de respuestas cualitativas válidas analizadas: {len(respuestas_p7)}\n")

# Mostrar las respuestas clave para el reporte
for i, respuesta in enumerate(respuestas_p7, 1):
    print(f"Respuesta {i}: {respuesta}")
    print("-" * 50)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Configuración estética de los gráficos
sns.set_theme(style="whitegrid")
plt.rcParams['figure.figsize'] = [10, 6]

# Cargar el archivo con el nombre exacto de tu formulario
df = pd.read_csv('Encuesta campo laboral en la comunicación digital - Respuestas de formulario 1.csv')

# Renombramos las columnas para facilitar el análisis en Python
df.columns = [
    'timestamp', 'orientacion', 'experiencia', 'rol', 'tipo_organizacion',
    'habilidades_importantes', 'comodidad_digital', 'desarrollo_habilidades',
    'p7_abierta', 'aspecto_satisfactorio', 'p9_desafios'
]

# Quitar espacios vacíos invisibles al inicio/final de los textos
for col in ['experiencia', 'rol', 'tipo_organizacion']:
    df[col] = df[col].str.strip()

print(f"¡Datos listos! Total de encuestados: {df.shape[0]}")
```

```python
# 1. Filtrar y limpiar las respuestas cualitativas reales
respuestas_filtradas = df['p7_abierta'].dropna()
# Descartamos respuestas vacías o que solo tengan un guion
respuestas_filtradas = respuestas_filtradas[respuestas_filtradas.str.strip().str.len() > 2]

print(f"Total de respuestas cualitativas válidas para analizar: {len(respuestas_filtradas)}\n")

# 2. Imprimir el corpus completo de respuestas para tu lectura y reporte de tesis
for i, respuesta in enumerate(respuestas_filtradas, 1):
    print(f"--- Respuesta #{i} ---")
    print(respuesta)
    print("\n" + "="*60 + "\n")
```

```python
keywords = {
    'Diseño / Canva / Gráfica': ['diseño', 'canva', 'gráfica', 'ilustrador', 'photoshop', 'adobe'],
    'Video / Audio / Edición': ['video', 'audio', 'edición', 'grabar', 'film', 'premiere', 'capcut'],
    'Redes Sociales / Community': ['redes', 'sociales', 'instagram', 'community', 'post', 'publicar'],
    'Inteligencia Artificial': ['ia', 'inteligencia artificial', 'chatgpt'],
    'Redacción / Copywriting / SEO': ['copy', 'copywriting', 'redacción', 'seo', 'escritura']
}

conteos_keywords = {k: 0 for k in keywords.keys()}
for res in respuestas_filtradas.str.lower():
    for categoria, palabras in keywords.items():
        if any(p in res for p in palabras):
            conteos_keywords[categoria] += 1

# Mostrar los resultados del minianálisis de texto
print("Frecuencia de temas mencionados en la aclaración de la P7:")
for cat, total in conteos_keywords.items():
    print(f"- {cat}: {total} menciones")
```

```python
# Definimos el orden lógico de los años de experiencia
orden_experiencia = ['Menos de 1 año', '1-3 años', '4-6 años', 'Más de 6 años']

# Calculamos e imprimimos los promedios
promedios_comodidad = df.groupby('experiencia')['comodidad_digital'].mean().reindex(orden_experiencia)
print("Promedio de Comodidad Digital por rango de experiencia:")
print(promedios_comodidad.round(2))

# Generamos el gráfico Boxplot
plt.figure(figsize=(9, 5))
sns.boxplot(data=df, x='experiencia', y='comodidad_digital', order=orden_experiencia, palette='Set2')
sns.stripplot(data=df, x='experiencia', y='comodidad_digital', order=orden_experiencia, color='black', alpha=0.4, jitter=True, size=6)

plt.title('¿A mayor experiencia existe mayor comodidad digital?', fontsize=13, pad=15)
plt.xlabel('Años de Experiencia', fontsize=11)
plt.ylabel('Nivel de Comodidad (Escala 1 al 5)', fontsize=11)
plt.ylim(0.5, 5.5)
plt.tight_layout()
plt.show()
```

```python
# Tabla de contingencia
tabla_roles_org = pd.crosstab(df['rol'], df['tipo_organizacion'])
print("\n--- Distribución de Roles por Tipo de Organización ---")
print(tabla_roles_org)

# Gráfico de Barras Apiladas
tabla_roles_org.plot(kind='bar', stacked=True, cmap='viridis', figsize=(11, 6))
plt.title('Estructura Organizacional según el Rol Desempeñado', fontsize=13, pad=15)
plt.xlabel('Rol o Posición Laboral', fontsize=11)
plt.ylabel('Cantidad de Profesionales', fontsize=11)
plt.xticks(rotation=20, ha='right')
plt.legend(title='Tipo de Organización', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
# Desglosamos las respuestas múltiples separadas por coma
desafios_desglosados = df['p9_desafios'].str.split(', ').explode().str.strip()

# Calculamos frecuencias absolutas y porcentajes (sobre el total de 37 encuestados)
conteos_desafios = desafios_desglosados.value_counts()
porcentajes_desafios = (conteos_desafios / len(df)) * 100

print("Frecuencia de los Desafíos Laborales en la Era Digital:")
for desafio, total in conteos_desafios.items():
    pct = (total / len(df)) * 100
    print(f"- {desafio}: {total} menciones ({pct:.1f}%)")

# Graficamos las respuestas múltiples de la Pregunta 9
plt.figure(figsize=(10, 5))
sns.barplot(y=conteos_desafios.index, x=conteos_desafios.values, palette='magma')
plt.title('Principales Desafíos del Profesional de la Comunicación en la Era Digital', fontsize=13, pad=15)
plt.xlabel('Cantidad de Respuestas', fontsize=11)
plt.ylabel('Desafíos de Calidad, Seguridad y Condiciones Empleo', fontsize=11)
plt.tight_layout()
plt.show()
```
