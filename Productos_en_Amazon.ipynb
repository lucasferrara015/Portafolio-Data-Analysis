```python
import pandas as pd
import glob
import os

ruta_carpeta = '/content/'


archivos_csv = glob.glob(os.path.join(ruta_carpeta, "*.csv"))


columnas_criticas = ['name', 'main_category', 'sub_category', 'actual_price', 'discount_price', 'ratings', 'no_of_ratings']

lista_dataframes = []

print(os.path.join(ruta_carpeta, "*.csv"))
print(f"Se encontraron {len(archivos_csv)} archivos para unificar.\n")

#  Bucle  para leer y procesar cada archivo
for archivo in archivos_csv:
    nombre_archivo = os.path.basename(archivo)
    try:
        # Leemos el CSV optimizando la carga usando solo las columnas críticas
        # errors='coerce' o encoding='utf-8' por si hay caracteres extraños de Amazon
        df_temporal = pd.read_csv(archivo, usecols=lambda c: c in columnas_criticas, low_memory=False)

        # Opcional: Si el CSV no tiene la columna 'main_category', le asignamos el nombre del archivo
        if 'main_category' not in df_temporal.columns or df_temporal['main_category'].isnull().all():
            df_temporal['main_category'] = nombre_archivo.replace('.csv', '').replace('All ', '')

        lista_dataframes.append(df_temporal)
        print(f"✔ Procesado con éxito: {nombre_archivo} ({len(df_temporal)} filas)")

    except Exception as e:
        print(f"❌ Error al leer {nombre_archivo}: {e}")

# CONCATENACIÓN: Unir todos los archivos verticalmente
print("\nUnificando todos los dataframes...")
df_amazon_maestro = pd.concat(lista_dataframes, ignore_index=True)

print(f"\n¡Estructura unificada con éxito!")
print(f"Total de registros combinados: {df_amazon_maestro.shape[0]} filas.")
print(f"Columnas conservadas: {list(df_amazon_maestro.columns)}")
```

```python
# Reemplaza 'actual_price' por el nombre real de columna de precio
# Eliminamos símbolos extraños y comas de miles, luego convertimos a numérico
df_amazon_maestro['price_clean'] = df_amazon_maestro['actual_price'].astype(str).str.replace('₹', '').str.replace('$', '').str.replace(',', '').str.strip()
df_amazon_maestro['price_clean'] = pd.to_numeric(df_amazon_maestro['price_clean'], errors='coerce')
```

```python
# Forzamos la columna ratings a tipo numérico. Los textos se convertirán en NaN automáticamente
df_amazon_maestro['ratings_clean'] = pd.to_numeric(df_amazon_maestro['ratings'], errors='coerce')
```

```python
# Eliminamos filas donde el precio limpio o el rating sean nulos
df_visual = df_amazon_maestro.dropna(subset=['price_clean', 'ratings_clean']).copy()
print(f"Filas listas para visualización después de eliminar nulos: {df_visual.shape[0]}")
```

```python
df_visual.to_csv('amazon_products_clean.csv', index=False)
```

```python
df_aires = df_visual[df_visual['main_category'].str.contains('appliances', case=False, na=False)]
```

```python
# Filtramos combinando la categoría principal y la subcategoría correcta
df_aires = df_visual[
    (df_visual['main_category'].str.contains('appliances', case=False, na=False)) &
    (df_visual['sub_category'].str.contains('Air Conditioners', case=False, na=False))
]

precio_medio = df_aires['price_clean'].mean()
precio_mediana = df_aires['price_clean'].median()

print("--- REPORTE DE PRECIOS DETALLADO: AIRES ACONDICIONADOS ---")
print(f"Total de modelos de Aires Acondicionados encontrados: {len(df_aires)}")
print(f"Precio Promedio (Media): ${precio_medio:.2f}")
print(f"Precio Central (Mediana): ${precio_mediana:.2f}")
```

```python
# Agrupamos por categoría principal y calculamos el precio máximo de cada una
top_categorias_caras = df_visual.groupby('main_category')['price_clean'].max().sort_values(ascending=False).head(5)

print("--- TOP 5 CATEGORÍAS PRINCIPALES MÁS CARAS ---")
print(top_categorias_caras)
```

```python
# Calculamos el coeficiente de correlación de Pearson
coef_correlacion = df_visual['price_clean'].corr(df_visual['ratings_clean'])

print("--- ANÁLISIS DE CORRELACIÓN GENERAL (Precio vs. Calificación) ---")
print(f"Coeficiente de Correlación de Pearson: {coef_correlacion:.4f}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load df_visual from the saved CSV to ensure it's defined
df_visual = pd.read_csv('amazon_products_clean.csv')

# Configuración estética profesional
sns.set_theme(style="whitegrid")
plt.rcParams['font.family'] = 'sans-serif'
plt.rcParams['figure.figsize'] = (16, 12)

# Crear un lienzo con 3 gráficos estructurados de nivel ejecutivo
fig, axes = plt.subplots(2, 2, figsize=(18, 14))
fig.suptitle('Amazon Products Market Analysis & Consumer Behavior', fontsize=22, fontweight='bold', color='#1a1a1a', y=0.96)

# -------------------------------------------------------------------------
# GRÁFICO 1: ¿Qué categorías contienen los productos con precios más altos?
# -------------------------------------------------------------------------
# Filtramos el outlier erróneo de 9.9 mil millones que descubrimos
df_filtrado_barras = df_visual[df_visual['price_clean'] < 1000000]
top_categorias = df_filtrado_barras.groupby('main_category')['price_clean'].max().sort_values(ascending=False).head(6).reset_index()

sns.barplot(
    x='price_clean', y='main_category', data=top_categorias,
    ax=axes[0, 0], palette='Oranges_r', hue='main_category', legend=False
)
axes[0, 0].set_title('Top 6 Categorías con los Precios Máximos Más Altos (Excl. Outliers)', fontsize=14, fontweight='bold', pad=10)
axes[0, 0].set_xlabel('Precio Máximo ($)', fontsize=11)
axes[0, 0].set_ylabel('Categoría Principal', fontsize=11)

# -------------------------------------------------------------------------
# GRÁFICO 2: ¿Existe una correlación entre precio y calificación?
# -------------------------------------------------------------------------
# Acotamos el precio a $150,000 para que la nube de puntos se expanda visualmente
df_filtrado_disp = df_visual[df_visual['price_clean'] < 150000]

# Tomamos una muestra aleatoria de 20,000 puntos para que el gráfico sea legible y estético
df_muestra = df_filtrado_disp.sample(n=min(20000, len(df_filtrado_disp)), random_state=42)

sns.scatterplot(
    x='price_clean', y='ratings_clean', data=df_muestra,
    ax=axes[0, 1], alpha=0.4, color='#FF9900', edgecolor='none'
)
# Dibujamos una línea de tendencia para demostrar matemáticamente la falta de correlación
sns.regplot(x='price_clean', y='ratings_clean', data=df_muestra, ax=axes[0, 1], scatter=False, color='#111111', line_kws={"linewidth": 2})

axes[0, 1].set_title('Análisis de Correlación: Precio vs. Calificación de Usuarios\n(Coeficiente de Pearson: -0.0043 → Correlación Nula)', fontsize=14, fontweight='bold', pad=10)
axes[0, 1].set_xlabel('Precio del Producto ($)', fontsize=11)
axes[0, 1].set_ylabel('Calificación (Estrellas)', fontsize=11)
axes[0, 1].set_ylim(0.8, 5.2)

# -------------------------------------------------------------------------
# GRÁFICO 3: Rango de precio dependiendo del tipo de subcategoría (Appliances)
# -------------------------------------------------------------------------
df_appliances_sub = df_visual[df_visual['main_category'].str.contains('appliances', case=False, na=False)]

# Usamos un Boxplot (Gráfico de Cajas) real para mostrar mínimos, máximos y cuartiles
sns.boxplot(
    x='price_clean', y='sub_category', data=df_appliances_sub,
    ax=axes[1, 0], palette='Set2', hue='sub_category', legend=False, showfliers=False
)
axes[1, 0].set_title('Distribución y Rangos de Precios por Subcategoría (Appliances)', fontsize=14, fontweight='bold', pad=10)
axes[1, 0].set_xlabel('Rango de Precios ($)', fontsize=11)
axes[1, 0].set_ylabel('Subcategoría', fontsize=11)

# -------------------------------------------------------------------------
# CUADRO INFORMATIVO: Reemplazo técnico de las tarjetas de KPI
# -------------------------------------------------------------------------
axes[1, 1].axis('off') # Ocultamos los ejes para usarlo como panel de texto ejecutivo
texto_kpi = (
    "KPI BENCHMARK:\n"
    "AIRES ACONDICIONADOS\n\n"
    f"• Total de Modelos Analizados: {len(df_visual[(df_visual['main_category'].str.contains('appliances', case=False, na=False)) & (df_visual['sub_category'].str.contains('Air Conditioners', case=False, na=False))])}\n"
    f"• Precio Promedio (Media): ${55042.30:,.2f}\n"
    f"• Precio Central (Mediana): ${56190.00:,.2f}\n\n"
    "INSIGHT CLAVE DEL PROYECTO:\n"
    "La simetría entre la Media y la Mediana en Climatización\n"
    "demuestra un mercado maduro y estandarizado. Al mismo tiempo,\n"
    "el coeficiente de Pearson cercano a cero (-0.0043) confirma que\n"
    "el precio en Amazon NO es un predictor de la satisfacción del cliente."
)
axes[1, 1].text(0.1, 0.2, texto_kpi, fontsize=13, fontweight='medium', color='#2c3e50', bbox=dict(facecolor='#f8f9fa', edgecolor='#e2e8f0', boxstyle='round,pad=1.5'))

# Ajustar diseño final y guardar
plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.savefig('amazon_portfolio_analytics.png', dpi=300, bbox_inches='tight')
plt.show()
```
